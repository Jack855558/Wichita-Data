```python
#get accident data
accident = pd.read_csv('Traffic_Accidents_2024.csv')
accident.tail(3)

# accident['DATE'] = pd.to_datetime(accident['DATE'])
```

```python

#handle missing values if necessiary
accident['CITY'] = accident['CITY'].fillna(accident['CITY'].mode()[0])
accident['STATE'] = accident['STATE'].fillna(accident['STATE'].mode()[0])


accident.loc[(accident['LOCATION'] == '1871 W SIM PARK DR') & (accident['POSTAL_CODE'].isna()), 'POSTAL_CODE'] = 67203.0
accident.loc[(accident['LOCATION'] == '4217 W 9TH ST') & (accident['POSTAL_CODE'].isna()), 'POSTAL_CODE'] = 67212.0

accident.isnull().sum()
```

```python
num_rows = accident.shape[0]
print(f"Number of rows in the DataFrame: {num_rows}")
```

```python
min_lat, max_lat = 37.5, 37.9  # Latitude bounds for Wichita
min_lon, max_lon = -97.5, -97.0  # Longitude bounds for Wichita

# Filter for rows that are outside the bounds
outside_data = accident[
    (accident['LATITUDE'] < min_lat) | (accident['LATITUDE'] > max_lat) | 
    (accident['LONGITUDE'] < min_lon) | (accident['LONGITUDE'] > max_lon)
]

# Drop rows that are outside the bounds
accident = accident.drop(outside_data.index)
print(f"Number of values outside the bounds: {outside_data.shape[0]}")
```

```python
print(weather['DATE'].duplicated().sum())

accident.head()
```

```python
#Edit weather data so that only max weather gets shown per day
weather_daily = weather.groupby('DATE', as_index=False).sum()

print(f"Total weather rows: {len(weather_daily)}")
print(f"Unique dates in weather: {weather_daily['DATE'].nunique()}")
weather_daily.head()
```

```python
accident.rename(columns={'ACCIDENT_DATE': 'DATE'}, inplace=True)

# Remove the timezone from the 'DATE' column (if it has timezone info)
# accident['DATE'] = accident['DATE'].dt.tz_localize(None)

#Remove time
accident['DATE'] = pd.to_datetime(accident['DATE'], format='%Y-%m-%d')


#Merge accident dates with weather dates
merged_df = pd.merge(accident, weather_daily, on='DATE', how='left')

# Total number of rows
total_rows = len(merged_df)

# Replace all NaN values with 0
merged_df = merged_df.fillna(0)

# Count the non-zero values in each column
prcp_non_zero = (merged_df['PRCP'] != 0).sum()
snow_non_zero = (merged_df['SNOW'] != 0).sum()
snwd_non_zero = (merged_df['SNWD'] != 0).sum()

# Print the results out of total rows
print(f"Non-zero count for Rain: {prcp_non_zero} out of {total_rows}")
print(f"Non-zero count for Snow: {snow_non_zero} out of {total_rows}")
print(f"Non-zero count for Snow depth: {snwd_non_zero} out of {total_rows}")


```
