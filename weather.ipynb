```python
import pandas as pd
```

```python
weather = pd.read_csv('weather.csv')
weather.head(10)
```

```python
#filter rows where name contians wichita
filtered_df = weather[weather['NAME'].str.contains('Wichita', case=False, na=False)]

filtered_df.head(10)   

# Optional: save the filtered data to a new CSV
filtered_df.to_csv('wichita_weather.csv', index=False)
```
