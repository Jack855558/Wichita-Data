```python
#Get data for streets, driving, biking, walk paths
import osmnx as ox

#Library for analyziing complex networks and graphs
import networkx as nx

#Converts address to latitude and longitude
from geopy.geocoders import Nominatim

#Interactive Graphs
import folium

#Static Graphing
import matplotlib.pyplot as plt

#Methods to get Geopy.geocoders to work
import ssl
import certifi
```

```python

#Download Road Data for Wichita, KS
G = ox.graph_from_place('Wichita, Kansas, USA', network_type='drive')
```

```python
start_node = ox.distance.nearest_nodes(G, X=-97.3308, Y=37.6922)  # Example coordinates
end_node = ox.distance.nearest_nodes(G, X=-97.3541, Y=37.6621)    # Example coordinates

# Find the shortest path by distance
shortest_route = nx.shortest_path(G, start_node, end_node, weight='length')

shortest_route
```

```python

# Create an SSL context that uses certifi's certificate bundle
ssl_context = ssl.create_default_context(cafile=certifi.where())

# Pass the ssl_context to the Nominatim geocoder
geolocator = Nominatim(user_agent="wichita_nav", ssl_context=ssl_context)
location = geolocator.geocode("123 Main St, Wichita, KS")
print(location.latitude, location.longitude)
```

```python
#Graph using OSMNX Comand
fig, ax = ox.plot_graph_route(G, shortest_route)
```
